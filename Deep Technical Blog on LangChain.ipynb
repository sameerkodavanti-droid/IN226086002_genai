```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(model="gpt-4o-mini")

print(llm.invoke("Explain LangChain in one sentence").content)
```

```python
from langchain.prompts import PromptTemplate

template = PromptTemplate.from_template(
    "Explain {topic} in simple terms"
)

chain = template | llm

print(chain.invoke({"topic": "LangChain"}).content)
```

```python
from langchain_core.runnables import RunnableSequence

chain = template | llm

result = chain.invoke({"topic": "AI Agents"})
print(result.content)
```

```python
from langchain.memory import ConversationBufferMemory

memory = ConversationBufferMemory()

memory.save_context({"input": "Hi"}, {"output": "Hello!"})
memory.save_context({"input": "My name is sameer"}, {"output": "Nice to meet you sameer"})

print(memory.load_memory_variables({}))
```

```python
from langchain.agents import initialize_agent, Tool

def multiply(a: int, b: int) -> int:
    return a * b

tools = [
    Tool(
        name="Multiplier",
        func=multiply,
        description="Multiply two numbers"
    )
]

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)

result = agent.invoke("What is 5 multiplied by 6?")
print(result)
```

```python
memory = ConversationBufferMemory()

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    memory=memory,
    verbose=True
)

agent.invoke("My name is Ravi")
agent.invoke("What is 10 * 5?")
response = agent.invoke("What is my name?")

print(response)
```

```python
from langchain_community.document_loaders import TextLoader

# Create file
with open("sample.txt", "w") as f:
    f.write("LangChain helps developers build AI applications.")

loader = TextLoader("sample.txt")
documents = loader.load()

print(documents)
```

```python
from langchain_community.vectorstores import FAISS
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings()

vectorstore = FAISS.from_documents(documents, embeddings)

query = "What is LangChain?"

docs = vectorstore.similarity_search(query)

print(docs[0].page_content)
```
