Task 1: Conceptual Understanding:

1. What is the difference between "Love" and "love" in NLP?

In raw NLP, "Love" and "love" are treated as two entirely different tokens because computers process text based on ASCII/Unicode character values. If left unstandardized, the model splits the frequency and context of the word, reducing its analytical power. Lowercasing ensures both are recognized as the same semantic entity.

2. What happens if stopwords are not removed?

Stopwords (like "is", "the", "and") appear with high frequency but carry very little contextual or semantic meaning. If they are not removed, they introduce noise into the dataset, causing the machine learning model to potentially give undue weight to statistically dominant but irrelevant words.

3. Provide two real-world scenarios where removing stopwords can be harmful.

Sentiment Analysis: Removing the stopword "not" from the phrase "I am not happy" completely flips the sentiment of the text to positive.

Search Engine Queries: If a user searches for a specific phrase like "To be or not to be," almost all the words are standard stopwords. Removing them would leave the search engine with no actionable input.

4. Explain the difference between stemming and lemmatization.

Stemming is a crude, rule-based process that simply chops off suffixes (e.g., "studying" becomes "studi"). It is fast but frequently results in non-dictionary root words.

Lemmatization uses morphological analysis and a dictionary to reduce a word to its proper base form or lemma (e.g., "better" becomes "good"). It is computationally heavier but highly accurate.


```python
#Task 2:Build Advanced Preprocessing Function:

import re

def reduce_repeated(word):
    # converts loooove -> love, goooood -> good
    return re.sub(r'(.)\1{2,}', r'\1', word)


def preprocess_text(text):

    if not isinstance(text, str):
        return [], ""

    text = text.strip()
    if text == "":
        return [], ""

    # lowercase
    text = text.lower()

    # remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)

    # remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # remove numbers
    text = re.sub(r'\d+', '', text)

    # remove special characters / emojis
    text = re.sub(r'[^a-z\s]', '', text)

    # remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # split words
    words = text.split()

    # handle repeated characters
    words = [reduce_repeated(word) for word in words]

    # remove small words (≤2) except "no", "not"
    tokens = []
    for word in words:
        if len(word) > 2 or word in ["no", "not"]:
            tokens.append(word)

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for sentence in test_sentences:
    tokens, clean_sentence = preprocess_text(sentence)

    print("Original:", sentence)
    print("Tokens  :", tokens)
    print("Cleaned :", clean_sentence)
    print("-" * 50)
```

```python
#Task 3: Stress Testing:

for i, sentence in enumerate(test_sentences, 1):

    tokens, cleaned_sentence = preprocess_text(sentence)

    print(f"Sentence {i}")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned_sentence)
    print("-" * 60)
```

```python
#Task 4: Token Analytics:

results = []

for sentence in test_sentences:

    tokens, cleaned_sentence = preprocess_text(sentence)

    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))

    if total_tokens > 0:
        avg_length = sum(len(word) for word in tokens) / total_tokens
    else:
        avg_length = 0

    results.append({
        "original": sentence,
        "tokens": tokens,
        "cleaned": cleaned_sentence,
        "total": total_tokens,
        "unique": unique_tokens,
        "avg_len": round(avg_length, 2)
    })
```

```python
for i, r in enumerate(results, 1):

    print(f"Sentence {i}")
    print("Original Text   :", r["original"])
    print("Tokens          :", r["tokens"])
    print("Cleaned Sentence:", r["cleaned"])
    print("Total Tokens    :", r["total"])
    print("Unique Tokens   :", r["unique"])
    print("Avg Token Length:", r["avg_len"])
    print("-" * 60)
```

Analysis Questions:

"Win $$$ now!!! Limited offer!!!" and "Worst service ever... 0/10" contained the highest volume of noise due to heavy use of special characters, symbols, and numbers.

Which sentence retained the most meaningful tokens after cleaning?

"I am not happy with this" retained excellent semantic meaning because the exception logic successfully preserved the word "not", keeping the sentiment intact.

```python
#Task 5: Frequency Analysis:

from collections import Counter

all_tokens = []

for r in results:
    all_tokens.extend(r["tokens"])
```

```python
token_counter = Counter(all_tokens)
```

```python
top_10 = token_counter.most_common(10)

print("Top 10 Most Frequent Words:")
for word, count in top_10:
    print(word, ":", count)
```

```python
least_5 = sorted(token_counter.items(), key=lambda x: x[1])[:5]

print("\nTop 5 Least Frequent Words:")
for word, count in least_5:
    print(word, ":", count)
```

```python
#Task 6: Build Full Pipeline:

def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned_sentence = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned_sentence)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
output = full_pipeline(test_sentences)

print("All Tokens:")
print(output["tokens"])

print("\nCleaned Sentences:")
for sentence in output["clean_sentences"]:
    print(sentence)
```
