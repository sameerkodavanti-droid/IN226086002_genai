###Section 3

###Task 1: Replace Hardcoded Prompts
In this task, we replace a simple hardcoded function with a reusable LangChain PromptTemplate. This removes the need for Python f-strings and keeps the logic entirely separate from the prompt text.

```python
# Define the template string
task1_template_str = "Explain {topic} in simple terms for beginners"

# Create a reusable PromptTemplate using input_variables
task1_prompt = PromptTemplate(
    input_variables=["topic"],
    template=task1_template_str
)

# Test the template
output_task1 = task1_prompt.format(topic="Large Language Models")
print("Task 1 Output:")
print(output_task1)
# Expected Output: Explain Large Language Models in simple terms for beginners
```

###Section 4

###Task 2: Multi-Input Prompt System
Here we build a single template that accepts three distinct inputs: topic, audience, and tone. We will test it against a provided list of test cases.

```python
# Define the multi-input template string
task2_template_str = "Explain {topic} for {audience} in a {tone} tone"

# Create the PromptTemplate
task2_prompt = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template=task2_template_str
)

# Define the test cases
test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

print("Task 2 Outputs:")
# Loop through test cases and format the prompt
for inputs in test_cases:
    formatted_prompt = task2_prompt.format(
        topic=inputs["topic"],
        audience=inputs["audience"],
        tone=inputs["tone"]
    )
    print(f"- {formatted_prompt}")
```

###Section 5

###Task 3: Prompt Variations Engine
This task requires us to create three different styles of prompts (Teaching, Interview, and Storytelling) and run them all for a single topic.

```python
# Define a dictionary holding our template variations
prompt_variations = {
    "Teaching": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} clearly step by step"
    ),
    "Interview": PromptTemplate(
        input_variables=["topic"],
        template="Ask 3 questions about {topic}"
    ),
    "Storytelling": PromptTemplate(
        input_variables=["topic"],
        template="Explain {topic} as a story"
    )
}

# Set the topic
target_topic = "Machine Learning"

print(f"Task 3 Outputs for topic: '{target_topic}'\n")

# Run the topic through all variations
for style, template in prompt_variations.items():
    formatted_variation = template.format(topic=target_topic)
    print(f"[{style} Style] -> {formatted_variation}")
```

###Section 6

###Task 4: ChatPromptTemplate System
For chat models, prompts are structured as a conversation between roles (e.g., System and User). We use ChatPromptTemplate to define the behavior dynamically via a role variable, and the message content via a topic variable.

```python
# Build a chat-based prompt system
chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a highly skilled {role}."),
    ("user", "Let's talk about {topic}.")
])

# Define our inputs
chat_role = "teacher"
chat_topic = "Neural Networks"

# Format the messages
chat_messages = chat_template.format_messages(role=chat_role, topic=chat_topic)

print("Task 4 Outputs (Formatted Message Objects):")
for message in chat_messages:
    # We display the type of message and its content
    message_type = type(message).__name__
    print(f"{message_type}: {message.content}")
```

###Section 7

###Task 5: Input Validation Layer
Robust prompt systems need input validation to ensure variables conform to expected constraints. Here, we validate audience and tone. If an invalid option is provided, we assign a safe default value.

```python
def validate_inputs(audience, tone):
    # Define acceptable values
    valid_audiences = ["beginner", "intermediate", "expert"]
    valid_tones = ["formal", "casual", "fun"]

    # Assign defaults if the input is invalid
    validated_audience = audience if audience in valid_audiences else "beginner"
    validated_tone = tone if tone in valid_tones else "casual"

    return validated_audience, validated_tone

print("Task 5 Validation Tests:")
# Test with valid inputs
print("Valid inputs (expert, fun):", validate_inputs("expert", "fun"))

# Test with invalid inputs to trigger default fallback
print("Invalid inputs (toddler, angry):", validate_inputs("toddler", "angry"))
```

###Section 8

###Task 6: Prompt Generator App
We will now combine everything into a function that validates inputs and then uses a PromptTemplate to generate the final prompt string without relying on Python f-strings.

```python
def generate_prompt(topic, audience, tone, style):
    # Step 1: Validate inputs
    v_audience, v_tone = validate_inputs(audience, tone)

    # Step 2: Define base template structure
    app_template_str = "Explain {topic} for {audience} in a {tone} {style} style"

    # Step 3: Create PromptTemplate
    app_prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone", "style"],
        template=app_template_str
    )

    # Step 4: Generate output
    return app_prompt.format(
        topic=topic,
        audience=v_audience,
        tone=v_tone,
        style=style
    )

print("Task 6 Output:")
# Test the prompt generator app
final_prompt = generate_prompt(
    topic="Neural Networks",
    audience="beginner",
    tone="fun",
    style="storytelling"
)
print(final_prompt)
# Expected Output: Explain Neural Networks for beginner in a fun storytelling style
```

###Section 9

###Task 7: Template Reusability Test
To prove the reusability of our system, we use a single PromptTemplate and loop over five completely different sets of inputs. The underlying structure stays identical, but the outputs dynamically adjust.

```python
# Create ONE template
reusable_template = PromptTemplate(
    input_variables=["concept", "analogy"],
    template="Explain the concept of {concept} by comparing it to {analogy}."
)

# 5 different input sets
test_inputs = [
    {"concept": "Cloud Computing", "analogy": "renting a storage unit"},
    {"concept": "APIs", "analogy": "a waiter in a restaurant"},
    {"concept": "Firewalls", "analogy": "a bouncer at a club"},
    {"concept": "Variables in programming", "analogy": "labeled boxes"},
    {"concept": "Data Encryption", "analogy": "a secret decoder ring"}
]

print("Task 7 Outputs (1 Template, 5 Variations):")
for inputs in test_inputs:
    print("- " + reusable_template.format(**inputs))
```

###Section 10

#Pipeline Summary
The workflow executed throughout this assignment models a standard real-world LLM data pipeline:

User Input: Raw data is captured from the user (e.g., topic, tone).

Validation Layer: Data is checked against strict rules, falling back to defaults to prevent bad prompts.

Prompt Template: The clean data is injected into a LangChain PromptTemplate or ChatPromptTemplate.

Dynamic Generation: A fully constructed prompt string or message array is securely formatted without f-strings.

Output: The prompt is ready to be sent to an LLM for inference.
